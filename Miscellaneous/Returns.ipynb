```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

jtplot.style(figsize=(15, 9))
```

```python
data = pd.read_csv('../data/BK_2002-01-01_2022-12-31_1d.csv', parse_dates=True, index_col=0)
price = data['Close'].to_frame()
```

$\displaystyle
r_t = \frac{p_t - p_0}{p_0}
\\
Simple\, Return = \frac{p_t}{p_0}
\\
Logarithmic\, Return = log(\frac{p_t}{p_0})
$

a positive log returns indicate growth, and negative log returns indicate loss


```python
price['pct_change'] = price['Close'].pct_change()                      # not additive
price['log_return'] = np.log(price['Close'] / price['Close'].shift(1)) # additive
```

```python
# price['pct_change'].cumsum().plot() USELESS!!!
price['log_return'].cumsum().plot()
```
