# Quantitative Momentum

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import yfinance as yf
import datetime as dt

import math
from scipy.stats import percentileofscore as score

from statistics import mean

from jupyterthemes import jtplot
jtplot.style(figsize=(15, 9))
```

```python
hqm_mom = pd.read_csv('../data/strategy/hqm_momentum.csv', index_col=0)
hqm_mom
```

```python
time_periods = [
    'One-Year',
    'Six-Month',
    'Three-Month',
    'One-Month'
]

for row in hqm_mom.index:
    for time_period in time_periods:
        change_col = f'{time_period} Price Return'
        percentile_col = f'{time_period} Return Percentile' 
        hqm_mom.loc[row, percentile_col] = score(hqm_mom[change_col], hqm_mom.loc[row, change_col])/100

hqm_mom
```

```python
for row in hqm_mom.index:
    momentum_percentiles = []
    for time_period in time_periods:
        momentum_percentiles.append(hqm_mom.loc[row, f'{time_period} Return Percentile'])
    hqm_mom.loc[row, 'HQM Scores'] = np.mean(momentum_percentiles)
hqm_mom
```

```python
hqm_mom.sort_values('HQM Scores', ascending=False, inplace=True)
hqm_mom = hqm_mom[:50]
hqm_mom.reset_index(inplace=True, drop=True)
hqm_mom
```

```python
portfolio_size = 10_000
```

```python
position_size = float(portfolio_size) / len(hqm_mom.index)
for i in hqm_mom.index:
    hqm_mom.loc[i, 'Number of Shares to Buy'] = math.floor(position_size/hqm_mom.loc[i, 'Price'])
hqm_mom
```
